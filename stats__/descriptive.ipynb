```python
import numpy as np
from scipy.stats import chi2, chisquare
import matplotlib.pyplot as plt

# Observed and expected frequencies
observed = np.array([60, 45, 45])  # Latte, Mocha, Cappuccino
expected = np.array([50, 50, 50])  # Equal distribution (150/3)

# Perform chi-squared goodness-of-fit test
chi2_stat, p_value = chisquare(observed, expected)

# Print results
print(f"Chi-Squared Statistic: {chi2_stat:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {len(observed) - 1}")

# Plot chi-squared distribution and test statistic
df = 5
x = np.linspace(0, 15, 1000)
pdf = chi2.pdf(x, df)

plt.figure(figsize=(8, 5))
plt.plot(x, pdf, 'b-', label=f'Chi-Squared PDF (df={df})')
plt.fill_between(x, pdf, where=(x >= chi2_stat), color='red', alpha=0.3, label='P-value region')
plt.axvline(chi2_stat, color='r', linestyle='--', label=f'Chi2 Stat = {chi2_stat:.2f}')
plt.title('Chi-Squared Distribution with Test Statistic')
plt.xlabel('Chi-Squared Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import threading

ITERS = 100000
x = [0]

def worker():
    for _ in range(ITERS):
        x[0] += 1  # this line creates a race condition
        # because it takes a value, increments and then writes
        # some inrcements can be done together, and lost

def main():
    x[0] = 0  # you may use `global x` instead of this list trick too
    t1 = threading.Thread(target=worker)
    t2 = threading.Thread(target=worker)
    t1.start()
    t2.start()
    t1.join()
    t2.join()

for i in range(5):
    main()
    print(f'iteration {i}. expected x = {ITERS*2}, got {x[0]}')
```
