```r
# Vectors to store ongoing data and sample means
xbars = c()
xbar = c()
observations = c()

# Draw a new observation, calculate mean, then store for plotting
for (n in 1:10000) {
  newdata = rcauchy(1) # Population with an undefined mean
  observations = c(observations, newdata)
  xbars = c(xbars, mean(observations))
}
```

```r
# Plot the observation values over time
plot(observations, type = "p", main = "Observation Values Over Time", xlab = "Iteration", ylab = "Observation Value")

# Plot the sample mean after each added value
plot(xbars, type = "p", main = "Sample Mean Over Time", xlab = "Iteration", ylab = "Sample Mean")
```
